## Exercice 1 (3 pts)

Écrire une fonction qui reçoit un nombre `x` et retourne :
- `'Positif'` si `x > 0`
- `'Négatif'` si `x < 0`
- `'Zéro'` sinon



```python
def verifier_nombre(x):
    if x > 0:
        return "Positif"
    elif x < 0:
        return "Négatif"
    else:
        return "Zéro"


# saisie utilisateur
x = float(input("Entrez un nombre : "))

# appel de la fonction
resultat = verifier_nombre(x)

print("Le nombre est :", resultat)
```

## Exercice 2 (3 pts)

Écrire une fonction qui reçoit un âge (entier) et retourne :
- `'Enfant'` si âge < 12
- `'Adolescent'` si 12 ≤ âge ≤ 17
- `'Adulte'` si âge ≥ 18



```python
def categorie_age(age):
    if age < 12:
        return "Enfant"
    elif age >= 12 and age <= 17:
        return "Adolescent"
    else:
        return "Adulte"


# saisie utilisateur
age = int(input("Entrez votre âge : "))

# appel de la fonction
resultat = categorie_age(age)

print("Catégorie :", resultat)
```

## Exercice 3 (3 pts)

Écrire une fonction qui reçoit une note sur 100 et retourne :
- `'Échec'` si note < 60
- `'Passable'` si 60–69
- `'Bien'` si 70–79
- `'Très bien'` si ≥ 80



```python

def evaluation_note(note):
    if note < 60:
        return "Échec"
    elif note >= 60 and note <= 69:
        return "Passable"
    elif note >= 70 and note <= 79:
        return "Bien"
    else:
        return "Très bien"


# saisie utilisateur
note = float(input("Entrez la note sur 100 : "))

# appel de la fonction
resultat = evaluation_note(note)

print("Résultat :", resultat)
```

## Exercice 4 (2 pts)

Écrire une fonction qui reçoit deux nombres `a` et `b` et retourne :
- le plus grand des deux
- ou la chaîne `'Égaux'` s’ils sont égaux



```python
def max_ou_egaux(a, b):
    if a > b:
        return a
    elif b > a:
        return b
    else:
        return "Égaux"


# saisie utilisateur
a = float(input("Entrez le premier nombre : "))
b = float(input("Entrez le deuxième nombre : "))

# appel de la fonction
resultat = max_ou_egaux(a, b)

print("Résultat :", resultat)


```

## Exercice 5 (4 pts)

Écrire une fonction qui reçoit un entier `n` et retourne une **liste de deux chaînes** :
1) `'Pair'` ou `'Impair'`
2) `'Multiple de 3'` ou `'Pas multiple de 3'`



```python
def analyser_nombre(n):
    if n % 2 == 0:
        parite = "Pair"
    else:
        parite = "Impair"

    if n % 3 == 0:
        multiple = "Multiple de 3"
    else:
        multiple = "Pas multiple de 3"

    return [parite, multiple]


# saisie utilisateur
n = int(input("Entrez un entier : "))

# appel de la fonction
resultat = analyser_nombre(n)

print(resultat)
```

## Exercice 7 (2 pts)

Écrire une fonction qui retourne la liste des nombres pairs entre 1 et 50 (inclus) avec une boucle `for`.



```python
def nombres_pairs():
    liste = []
    for i in range(1, 51):
        if i % 2 == 0:
            liste.append(i)
    return liste
resultat = nombres_pairs()
print(resultat)
```

## Exercice 12 (2 pts)

Écrire une fonction qui reçoit une liste de nombres saisis successivement et retourne la somme jusqu’à rencontrer un 0.

Exemple : `[5, 2, -1, 0, 100]` -> somme = `6`.

Consigne : utiliser une boucle `while`.




```python
def somme_jusqua_zero(liste):
    somme = 0
    i = 0

    while i < len(liste) and liste[i] != 0:
        somme += liste[i]
        i += 1

    return somme
nombres = [5, 2, -1, 0, 100]
resultat = somme_jusqua_zero(nombres)
print("Somme =", resultat)


```
